```python
def percentage_to_float(percentage_str):
    if percentage_str == 'Don\'t Want to Reveal':
        return None
    elif percentage_str.startswith('Above'):
        return float(percentage_str.split(' ')[1].replace('%', ''))
    elif percentage_str.startswith('Upto'):
        return float(percentage_str.split(' ')[1].replace('%', ''))
    else:
        bounds = percentage_str.split(' to ')
        return (float(bounds[0].replace('%', '')) + float(bounds[1].replace('%', ''))) / 2

# Apply the custom function to convert the 'Percentage of Investment' column
df['Percentage of Investment'] = df['Percentage of Investment'].apply(percentage_to_float)

# If you want to print all the converted values
for index, value in enumerate(df['Percentage of Investment']):
    print(f"Index: {index}, Float Value: {value}")
```

```python
def household_income_to_float(income_str):
    if income_str == 'Don\'t Want to Reveal':
        return None
    elif income_str.startswith('Above'):
        return float(income_str.split(' ')[2].replace('US$', '').replace(',', ''))
    elif income_str.startswith('Upto'):
        return float(income_str.split(' ')[2].replace('US$', '').replace(',', ''))
    elif ' to ' in income_str:
        bounds = income_str.split(' to ')
        return (float(bounds[0].replace('US$', '').replace(',', '')) + float(bounds[1].replace('US$', '').replace(',', ''))) / 2
    else:
        return float(income_str.replace('US$', '').replace(',', ''))

# Apply the custom function to convert the 'Household Income' column
df['Household Income'] = df['Household Income'].apply(household_income_to_float)

# If you want to print all the converted values
for index, value in enumerate(df['Household Income']):
    print(f"Index: {index}, Float Value: {value}")
```

```python
def identify_strong_correlations(correlation_matrix, threshold=0.8):
    strong_correlations = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            if abs(correlation_matrix.iloc[i, j]) >= threshold:
                strong_correlations.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))
    return strong_correlations

# Use the function to identify strong correlations
strong_correlations = identify_strong_correlations(correlation_matrix, threshold=0.8)

# Print the identified strong correlations
for correlation in strong_correlations:
    print(correlation)
```

# Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
numerical_columns = ['Number of investors in family', 'Knowledge level about different investment product',
                 'Knowledge level about sharemarket', 'Knowledge about Govt. Schemes', 'Household Income', 'Percentage of Investment']
categorical_columns = ['Age', 'Role', 'Investment Influencer', 'Gender', 'Education', 'City',
                   'Reason for Investment',
                   'Investment Experience', 'Source of Awareness about Investment', 'Marital Status']

df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

encoder = OneHotEncoder(drop='first')
encoded_categorical = pd.DataFrame(encoder.fit_transform(df[categorical_columns]).toarray(),columns=encoder.get_feature_names_out(categorical_columns))
scaler = StandardScaler()
scaled_numerical = scaler.fit_transform(df[numerical_columns])
scaled_numerical = pd.DataFrame(scaled_numerical, columns=numerical_columns)
data = pd.concat([encoded_categorical, scaled_numerical], axis=1)
```

```python
data
```

# Label Encoding target variables

```python
label_encoder_risk = LabelEncoder()
label_encoder_return = LabelEncoder()
outcome_encoded_risk = label_encoder_risk.fit_transform(df['Risk Level'])
outcome_encoded_return = label_encoder_return.fit_transform(df['Return Earned'])
outcome_encoded = np.column_stack((outcome_encoded_risk, outcome_encoded_return))
```

# Random Forest for Feature Selection

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(data, outcome_encoded)
feature_importances = rf_classifier.feature_importances_
sorted_indices = np.argsort(feature_importances)[::-1]
top_features = sorted_indices[:10]
selected_features = data.columns[top_features]
plt.figure(figsize=(10, 6))
plt.bar(range(len(selected_features)), feature_importances[top_features], color='skyblue') 
plt.xticks(range(len(selected_features)), selected_features, rotation=45, ha='right') 
plt.ylabel('Feature Importance')  
plt.xlabel('Feature')  
plt.title('Feature Importance Plot')  
plt.gca().invert_yaxis()
plt.show()
```

```python
selected_features
```

# Train & Test Data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

X_selected = data[selected_features]
X_train, X_test, y_train, y_test = train_test_split(X_selected, outcome_encoded, test_size=0.2, random_state=42)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
target_accuracies = [accuracy_score(y_test[:, i], predictions[:, i]) for i in range(y_test.shape[1])]
target_classification_reports = []
target_variable_names = ['Risk Level', 'Return Earned']
for i, target_name in enumerate(target_variable_names):
    report = classification_report(y_test[:, i], predictions[:, i])
    target_classification_reports.append((target_name, report))
print(f"Overall Accuracy - {np.mean(target_accuracies):.4f}")
for target_name, report in target_classification_reports:
    print(f"\nClassification Report for {target_name}:\n{report}")
```

```python
def encodeInput(input_data, category_values):
    encoded_columns = []
    for col, values in category_values.items():
        for val in values:
            encoded_columns.append(f"{col}_{val}")
    input_encoded = pd.DataFrame(columns=encoded_columns)
    for idx, row in input_data.iterrows():
        encoded_row = []
        for col, values in category_values.items():
            for val in values:
                if row[col] == val:
                    encoded_row.append(1)
                else:
                    encoded_row.append(0)
        input_encoded.loc[idx] = encoded_row
    return input_encoded
```

```python
def predict(model, X_train, label_encoder_risk, label_encoder_return, dropdowns, output_frame, categorical_columns, category_values):
    target_variable_names = ['Risk Level', 'Return Earned']
    user_inputs = {}
    for col in categorical_columns:
        user_inputs[col] = dropdowns[col].get()
    
    new_input = pd.DataFrame(user_inputs, index=[0])
    new_input_encoded = encodeInput(new_input, category_values)
    new_input_encoded = new_input_encoded.reindex(columns=X_train.columns, fill_value=0)
    new_predictions = model.predict(new_input_encoded)
    decoded_predictions = np.column_stack((label_encoder_risk.inverse_transform(new_predictions[:, 0]), label_encoder_return.inverse_transform(new_predictions[:, 1])))

    for i, target_name in enumerate(target_variable_names):
        prediction_label = ttk.Label(output_frame, text=f"{target_name}: {decoded_predictions[0, i]}")
        prediction_label.grid(row=i, column=0, sticky="w")
```

# Creating an application

```python
import tkinter as tk
from tkinter import ttk

# Function to center the window on the screen
def center_window(window, width, height):
    screen_width = window.winfo_screenwidth()
    screen_height = window.winfo_screenheight()
    x = (screen_width - width) // 2
    y = (screen_height - height) // 2
    window.geometry(f"{width}x{height}+{x}+{y}")

# Function to update window size based on content
def update_window_size(window):
    window.update_idletasks()
    window_width = window.winfo_width()
    window_height = window.winfo_height()
    center_window(window, window_width, window_height)

# Creating a Tkinter window
window = tk.Tk()
window.title("Predictor")

# Defining categories for dropdowns in UI
category_values = {}
for i in ['Age', 'Role', 'Investment Influencer', 'Gender', 'Education', 'City', 'Reason for Investment', 'Household Income', 'Percentage of Investment', 'Investment Experience', 'Source of Awareness about Investment', 'Marital Status']:
    category_values[i] = df[i].unique()

# Order in which inputs should be displayed
input_order = ['Age', 'Role', 'Investment Influencer', 'Gender', 'Education', 'City', 'Reason for Investment', 'Household Income', 'Percentage of Investment', 'Investment Experience', 'Source of Awareness about Investment', 'Marital Status']

# Creating a frame for input widgets
input_frame = ttk.Frame(window)
input_frame.pack(padx=20, pady=10)

# Creating dropdowns for each feature in the specified order
inputs = {}
columns_to_exclude = ['Risk Level', 'Return Earned', 'S. No.']
for i, column in enumerate(input_order):
    if column not in columns_to_exclude:
        label = ttk.Label(input_frame, text=column)
        label.grid(row=i, column=0, padx=5, pady=5, sticky="w")
        ip = ttk.Combobox(input_frame, values=df[column].unique().tolist(), state="readonly")
        ip.grid(row=i, column=1, padx=5, pady=5, sticky="ew")
        ip.current(0)
        inputs[column] = ip

# Creating a frame for output labels
output_frame = ttk.Frame(window)
output_frame.pack(padx=20, pady=10)

# Button to trigger the prediction
def on_predict():
    predict(model, X_train, label_encoder_risk, label_encoder_return, inputs, output_frame, ['Age', 'Role', 'Investment Influencer', 'Gender', 'Education', 'City', 'Reason for Investment', 'Household Income', 'Percentage of Investment', 'Investment Experience', 'Source of Awareness about Investment', 'Marital Status'], category_values)
    update_window_size(window)

predict_button = ttk.Button(window, text="Get Prediction", command=on_predict)
predict_button.pack(padx=20, pady=10)

# Center the window on the screen
center_window(window, 550, 500)  # Initial window size
window.configure(bg='lightblue')
window.mainloop()
```
